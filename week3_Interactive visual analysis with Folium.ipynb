```python
# TASK 3: Calculate the distances between a launch site to its proximities
import numpy as np

# Function to calculate the Haversine distance
def haversine(lat1, lon1, lat2, lon2):
    # Radius of Earth in kilometers
    R = 6371.0
    
    # Convert latitudes and longitudes from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Differences in coordinates
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Haversine formula
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    # Distance in kilometers
    distance = R * c
    return distance

# Example launch sites with their latitude and longitude
launch_sites = {
    'CCAFS SLC-40': (28.563197, -80.576820),
    'KSC LC-39A': (28.573255, -80.646895),
    'VAFB SLC-4E': (34.632834, -120.610746),
    'Site D': (51.423805, -0.334659)
}

# Reference launch site (e.g., CCAFS SLC-40)
ref_site = 'CCAFS SLC-40'
ref_lat, ref_long = launch_sites[ref_site]

# Calculate distances to other launch sites
distances = {}
for site, (lat, long) in launch_sites.items():
    if site != ref_site:  # Exclude the reference site itself
        distance = haversine(ref_lat, ref_long, lat, long)
        distances[site] = distance

# Print distances
for site, distance in distances.items():
    print(f"Distance from {ref_site} to {site}: {distance:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
